```python
def feature_class_to_pandas_data_frame(feature_class, field_list):
    """
    Produce a pandas dataframe from a feature class. Requires a list 
    of fields to import. Can specify how to treat NULL values.
    """
    return pd.DataFrame(
		arcpy.da.FeatureClassToNumPyArray(
			in_table=feature_class,
			field_names=field_list,
			skip_nulls=False,
			null_value=0
		),
        columns=field_list
	)
```
